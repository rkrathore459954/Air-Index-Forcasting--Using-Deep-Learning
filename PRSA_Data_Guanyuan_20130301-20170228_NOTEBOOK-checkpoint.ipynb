```python
# outlier detection 
# Method used Z-score, IQR, 
#
dropped = new_data.drop(['station'], axis=1)
from scipy import stats
z = np.abs(stats.zscore(dropped)) <2
z
```

```python
#Z-score
data_Z = dropped[z.all(axis=1)]
data_Z.shape
```

```python
#IQR 
q1 = np.percentile(new_data, 25, method = "midpoint")
q3 = np.percentile(new_data, 75, method = "midpoint")
iqr = q3 -q1
```

```python
#IQR(removal)

Q1 = dropped.quantile(0.25)
Q3 = dropped.quantile(0.75)
IQR = Q3-Q1
data_IQR = dropped[~((dropped<(Q1-1.5*IQR)) | (dropped> (Q3+1.5*IQR))).any(axis=1)]
data_IQR.shape
```

```python
#replacing


median = dropped.median()
std = dropped.std()
outliers = (dropped-median).abs() > std
dropped[outliers] = np.nan
data_replace = dropped.fillna(median)
data_replace
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from keras.models import Sequential
from keras.layers import LSTM, Dense, Conv1D, Flatten
from keras.optimizers import Adam

# Step 1: Data Preprocessing
# Assuming your features are stored in 'X' and the target variable in 'y'
# Perform any necessary cleaning and normalization steps

X_scaled = data_replace[cor_features]
target_values = data_replace['PM2.5']

# Step 2: Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, target_values, test_size=0.2, random_state=42)

# Step 3: Model Training - LSTM
lstm_model = Sequential()
lstm_model.add(LSTM(64, input_shape=(X_train.shape[1], 1)))
lstm_model.add(Dense(1))
lstm_model.compile(loss='mean_squared_error', optimizer=Adam())
lstm_model.fit(np.expand_dims(X_train, axis=2), y_train, epochs=100, batch_size=50, verbose=0)

# Step 4: Model Training - CNN
cnn_model = Sequential()
cnn_model.add(Conv1D(32, kernel_size=3, activation='relu', input_shape=(X_train.shape[1], 1)))
cnn_model.add(Flatten())
cnn_model.add(Dense(1))
cnn_model.compile(loss='mean_squared_error', optimizer=Adam())
cnn_model.fit(np.expand_dims(X_train, axis=2), y_train, epochs=100, batch_size=32, verbose=0)

# Step 5: Model Integration - Averaging Predictions
lstm_pred = lstm_model.predict(np.expand_dims(X_test, axis=2))
cnn_pred = cnn_model.predict(np.expand_dims(X_test, axis=2))
hybrid_pred = np.mean([lstm_pred.flatten(), cnn_pred.flatten()], axis=0)

# Step 6: Model Evaluation
mse = mean_squared_error(y_test, hybrid_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, hybrid_pred)
r2 = r2_score(y_test, hybrid_pred)

print("Hybrid Model Performance:")
print("MSE:", mse)
print("RMSE:", rmse)
print("MAE:", mae)
print("R-squared:", r2)
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from keras.models import Model
from keras.layers import Input, Dense, Lambda
from keras import backend as K
from keras import metrics


# Preprocess the data
target_column = 'PM2.5'  # Replace 'target' with the actual column name
target_values = data_replace[target_column].values
data = data_replace.drop(columns=[target_column])
X_train, X_test, y_train, y_test = train_test_split(data, target_values, test_size=0.2, shuffle=False)

# Normalize the input features
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Define the VAE architecture
input_dim = X_train.shape[1]
latent_dim = 2  # Set the desired latent dimension

# Encoder architecture
inputs = Input(shape=(input_dim,))
enc_layer = Dense(64, activation='relu')(inputs)
z_mean = Dense(latent_dim)(enc_layer)
z_log_var = Dense(latent_dim)(enc_layer)

# Sampling layer
def sampling(args):
    z_mean, z_log_var = args
    epsilon = K.random_normal(shape=(K.shape(z_mean)[0], latent_dim), mean=0., stddev=1.)
    return z_mean + K.exp(z_log_var / 2) * epsilon

z = Lambda(sampling)([z_mean, z_log_var])

# Decoder architecture
dec_layer = Dense(64, activation='relu')(z)
outputs = Dense(1)(dec_layer)

# Create the VAE model
model = Model(inputs, outputs)

# Define the loss function and the KL divergence regularization term
reconstruction_loss = metrics.mean_squared_error(inputs, outputs)
kl_loss = -0.5 * K.sum(1 + z_log_var - K.square(z_mean) - K.exp(z_log_var), axis=-1)
vae_loss = K.mean(reconstruction_loss + kl_loss)

# Compile the model
model.add_loss(vae_loss)
model.compile(optimizer='adam')

# Train the VAE model
model.fit(X_train, epochs=50, batch_size=32, verbose=0)

# Generate predictions on the test set
predictions = model.predict(X_test)

# mse = mean_squared_error(y_test, predictions)
# rmse = np.sqrt(mse)
# mae = mean_absolute_error(y_test, predictions)
# r2 = r2_score(y_test, predictions)

# print("VAE Model Performance:")
# print("MSE:", mse)
# print("RMSE:", rmse)
# print("MAE:", mae)
# print("R-squared:", r2)
```
